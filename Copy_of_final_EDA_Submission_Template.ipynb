## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
# Stays above 1 day and less than 3 days in weekend nights from customer.
a[(a['stays_in_weekend_nights'] > 1) & (a['stays_in_weekend_nights'] < 3)]

```

```python
# Creating column for total arrival by adding weekday and weekend stay in night
a['Total_stay_night'] = a['stays_in_week_nights'] + a['stays_in_weekend_nights']
# Creating total People stay in a hotel including children , Adults and Babies
a['Total_people_stay'] = a['adults'] + a['babies'] + a['children']
```

```python
# Creating a column on ratings based upon repeated guest
def a_condition(ratings):
  if ratings['is_repeated_guest'] > 0:
    return 5
  else:
    return 0
a['new_rating'] = a.apply(a_condition, axis=1)
print(a)
# if ratings are 5 then there are repeated guest visiting hotels. Hence we have used apply function.
```

```python
a.columns
```

```python
# total number of special request in each hotel
a[a['total_of_special_requests'] > 0]
# printing total number of special request above 0.
```

```python
#Finding maximum total number of special request in a year from hotels and resorts.
a.groupby('arrival_date_year')['total_of_special_requests'].max()
```

```python
#Finding maximum number of total stay at night from hotels and resorts.
a.groupby('arrival_date_year')['Total_stay_night'].max()
```

```python
#Finding sum of total number of people stays in hotel and resorts in respective to years
a.groupby('arrival_date_year')['Total_people_stay'].sum()
```

```python
#Finding mean of total number of people stays in hotel and resorts in respective to years
a.groupby('arrival_date_year')['Total_people_stay'].mean()
```
